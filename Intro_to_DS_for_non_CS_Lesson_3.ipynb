```python
import json

countries_file = open('countries.json')
county_codes = json.load(countries_file)

print(county_codes)
```

```python
def get_country_name(code):
  code = code.upper()
  if code in county_codes:
    return county_codes[code]
  else:
    return code
```

```python
for line in data[1:]:
  line_formatted = line.strip().split(",")
  year = int(line_formatted[0])
  from_c = get_country_name(line_formatted[2])
  to_c = get_country_name(line_formatted[3])
  points = int(line_formatted[6])

  if year >= 1990 and year <= 2003:
    if points == 12 and to_c == "Cyprus":
      print(from_c, "gave", to_c, points, "points in", year)
```

```python
from PIL import Image

import glob
flag_files = glob.glob("flags/*.png")

print(flag_files)
```

```python
for filename in sorted(flag_files):
  white_px = 0
  rgb_flag_image = Image.open(filename).convert('RGB')
  w, h = rgb_flag_image.size
  for x in range(w):
    for y in range(h):
      if rgb_flag_image.getpixel((x, y)) == (255, 255, 255):
        white_px += 1

  white_percent = round(white_px * 100 / (w * h), 2)
  print(filename, "\t", white_percent)
```
