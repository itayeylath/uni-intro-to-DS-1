```python
my_grades = [89, 95, 86, 95, 100, 89] # List of my grades
print(my_grades)

my_min_grade = min(my_grades)
my_max_grade = max(my_grades)
print("Min:", my_min_grade, " Max:", my_max_grade)

my_average_grade = round(sum(my_grades)/len(my_grades), 2)
print("My average is", my_average_grade)
```

```python
visits = {"Q2": 8, "Q1": 15, "Q3": 14, "Q4": 12}
print(visits)

for v in visits:
  print(v, visits[v])
```

```python
visits = {"Q2": 8, "Q1": 15, "Q3": 14, "Q4": 12}
print(visits)

for v in sorted(visits):
  print(v, visits[v])
```

```python
visits = {"Q1": 15, "Q2": 8, "Q3": 14, "Q4": 12}
print(visits)

for v in sorted(visits, key=lambda x: visits[x], reverse=True):
  print(v, visits[v])
```

```python
city = "New York"
stadium_name = "Madison Square Garden"
full_address = "4 Pennsylvania Plaza, New York, USA"

if city in full_address:
  print(stadium_name, "is in", city)
else:
  print(stadium_name, "is NOT in", city)
```

```python
full_address = "4 Pennsylvania Plaza, New York, USA"

counter_a = 0
counter_N = 0
for ch in full_address:
  if ch == "a":
    counter_a = counter_a + 1
  if ch == "N":
    counter_N += 1

print("The letter 'a' appeared", counter_a, "times")
print("The letter 'N' appeared", counter_N, "times")
```
