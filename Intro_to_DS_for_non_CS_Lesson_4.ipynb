```python
# Working with "Billboard "The Hot 100" Songs" dataset
# https://www.kaggle.com/datasets/dhruvildave/billboard-the-hot-100-songs

import pandas as pd

charts = pd.read_csv("charts.csv")
display(charts.head(20))
```

```python
print(charts.shape)
print(charts.columns)
print(charts.info())
```

```python
display(charts[(charts['artist'] == 'Adele')].head(30))
```

```python
adele_songs = charts[(charts['artist'] == 'Adele')]

print(adele_songs.shape)
display(adele_songs.head())
```

```python
display(adele_songs[(adele_songs['rank'] == 1)])
```

```python
display(adele_songs[(adele_songs['rank'] == 1) &
                    (adele_songs['last-week'].isnull())])
```

```python
display(adele_songs[(adele_songs['rank'] <= 10) &
                    (adele_songs['last-week'].isnull())])
```

```python
display(adele_songs[adele_songs['weeks-on-board'] == 1])
```

```python
print(adele_songs['weeks-on-board'].max())
```

```python
max_weeks = adele_songs['weeks-on-board'].max()
display(adele_songs[adele_songs['weeks-on-board'] == max_weeks])
```

```python
display(adele_songs[(adele_songs['song'] == 'Rolling In The Deep')])
```

```python
print(charts.groupby(['date'])['date'].count())
```

```python
songs = charts[charts['weeks-on-board'] == 1]
songs.groupby(['artist'])['artist'].count().sort_values(ascending=False).head(20)
```

```python
weeknd = charts[charts['artist'].str.contains("The Weeknd")]
display(weeknd)
```

```python
weeknd = weeknd[weeknd['weeks-on-board'] == 1]
display(weeknd)
```

```python
weeknd = weeknd.drop(columns=['last-week', 'weeks-on-board', 'peak-rank'])
display(weeknd)
```

```python
weeknd['feat'] = weeknd['artist'].str.replace("The Weeknd", "")
display(weeknd)
```

```python
display(weeknd[weeknd['feat'].str.len() > 1])
```

```python
weeknd['feat'] = weeknd['feat'].str.replace("&", "")
weeknd['feat'] = weeknd['feat'].str.replace("Featuring", "")
weeknd['feat'] = weeknd['feat'].str.replace(" X ", "")
display(weeknd[weeknd['feat'].str.len() > 1])
```

```python
weeknd['year'] = weeknd['date'].str[0:4]
print(weeknd.groupby('year')['year'].count())
```

```python
weeknd['month'] = weeknd['date'].str[5:7]
print(weeknd.groupby('month')['month'].count())
```

```python
weeknd['day'] = weeknd['date'].str[8:10]
print(weeknd.groupby('day')['day'].count())
```
