```python
# Quastion 1
import requests

# Fetch data from the API
url = "http://api.open-notify.org/astros.json"
response = requests.get(url)
data = response.json()

num_people = data['number']
print(f"num of pepole in space {num_people}")

crafts = set(person['craft'] for person in data['people'])
num_crafts = len(crafts)
print(f"num of crafts in space {num_crafts}")

```

```python
# Question 2
import pandas as pd
from collections import Counter

url = "https://gist.githubusercontent.com/namuol/2657233/raw/74135b2637e624848c163759be9cd14ae33f5153/presidents.csv"
df = pd.read_csv(url)

df.columns = df.columns.str.strip()

state_counts = df['Home State'].value_counts()
top_state = state_counts.idxmax()
print(f"1. State with most presidents: {top_state}")

first_names = df['President'].apply(lambda x: x.split()[0])
name_counts = Counter(first_names)
most_common_name, count = name_counts.most_common(1)[0]
print(f"2. Most common first name: {most_common_name} ({count} presidents)")
```

```python
# Question 3 - Setup
import json
import glob
from PIL import Image
from collections import Counter

# Load country codes
with open('../DataForLesson3/countries.json') as f:
    country_codes = json.load(f)

def get_country_name(code):
    code = code.upper()
    if code in country_codes:
        return country_codes[code]
    return code

# Get all flag files
flag_files = glob.glob('../DataForLesson3/flags/*.png')
print(f"Found {len(flag_files)} flags")
```

```python
# Question 3.1
square_flags = []
for filename in sorted(flag_files):
    img = Image.open(filename)
    w, h = img.size
    
    if w == h:
        country_code = filename.split("/")[-1].split(".")[0]
        country_name = get_country_name(country_code)
        square_flags.append(country_name)
        print(f"{country_name} ({w}x{h})")

print(f"\nTotal: {len(square_flags)} countries with square flags")
```

```python
# Question 3.2
print("Countries with exactly 3 colors in their flag:")
print("-" * 40)

three_color_flags = []
for filename in sorted(flag_files):
    img = Image.open(filename).convert('RGB')
    w, h = img.size
    
    colors = set()
    for x in range(w):
        for y in range(h):
            colors.add(img.getpixel((x, y)))
    
    if len(colors) == 3:
        country_code = filename.split("/")[-1].split(".")[0]
        country_name = get_country_name(country_code)
        three_color_flags.append(country_name)
        print(f"{country_name} - colors: {colors}")

print(f"\nTotal: {len(three_color_flags)} countries with exactly 3 colors")
```

```python
# Question 3.3
print("Finding the 3 most common colors across all flags...")
print("-" * 40)

all_colors = Counter()

for filename in flag_files:
    img = Image.open(filename).convert('RGB')
    w, h = img.size
    
    for x in range(w):
        for y in range(h):
            color = img.getpixel((x, y))
            all_colors[color] += 1

top_3_colors = all_colors.most_common(3)

print("\nTop 3 most common colors:")
for i, (color, count) in enumerate(top_3_colors, 1):
    print(f"{i}. RGB{color} - appeared {count:,} times")
```
